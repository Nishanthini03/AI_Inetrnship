**Linear algebra, probability, and calculus for AI model development.**

```python
import numpy as np # For numerical operations
vector = np.array([1, 2, 3])  # 1D
matrix = np.array([[1, 2], [3, 4]])  # 2D
tensor = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])  # 3D

print("Vector:\n", vector)
print(f"Matrix:\n{matrix}\n")
print(f"Tensor:\n{tensor}\n")
```

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

sum_matrix = A + B
prod_matrix = A @ B  # Matrix multiplication
transposed = A.T  # Transpose
inverse_A = np.linalg.inv(A)  # Inverse of A

print("Sum of Matrices:\n", sum_matrix)
print("Matrix Product:\n", prod_matrix)
print("Transpose of A:\n", transposed)
print("Inverse of A:\n", inverse_A)
```

```python
v1 = np.array([2, 3, 4])
v2 = np.array([1, 0, -1])
dot_product = np.dot(v1, v2)
print("Dot Product:", dot_product)
```

```python
eigenvalues, eigenvectors = np.linalg.eig(A)

print("Eigenvalues:\n", eigenvalues)
print("Eigenvectors:\n", eigenvectors)
```

```python
import matplotlib.pyplot as plt
from scipy.stats import bernoulli

p = 0.6  # Probability of success
samples = bernoulli.rvs(p, size=1000)  # Generate 1000 samples

# Plot histogram
plt.hist(samples, bins=2, density=True, alpha=0.6, color='b', edgecolor='black')
plt.xticks([0, 1], ['0 (Failure)', '1 (Success)'])
plt.xlabel("Outcome")
plt.ylabel("Probability")
plt.title("Bernoulli Distribution (p=0.6)")
plt.show()
```

```python
from scipy.stats import binom

n = 10  # Number of trials
p = 0.6  # Probability of success
x = np.arange(0, n+1)  # Possible outcomes (0 to 10)
pmf = binom.pmf(x, n, p)  # Probability mass function

# Plot
plt.bar(x, pmf, color='b', edgecolor='black', alpha=0.7)
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.title(f"Binomial Distribution (n={n}, p={p})")
plt.show()
```

```python
from scipy.stats import multinomial

n = 10  # Total trials
p = [1/6]*6  # Probabilities for a fair die

samples = multinomial.rvs(n, p, size=1000)  # Generate 1000 samples

# Plot histogram of one experiment
plt.bar(range(1, 7), samples[0], color='purple', edgecolor='black', alpha=0.7)
plt.xlabel("Die Face")
plt.ylabel("Count")
plt.title(f"Multinomial Distribution (Rolling a Die {n} times)")
plt.xticks(range(1, 7))
plt.show()
```

```python
import scipy.stats as stats # For statistical functions
import matplotlib.pyplot as plt # For plotting

x = np.linspace(-3, 3, 100) # Generate 100 points evenly spaced between -3 and 3
y = stats.norm.pdf(x, 0, 1)  # Mean=0, Std=1

plt.plot(x, y)
plt.title("Gaussian Distribution")
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.grid()
plt.show()
```

```python
from scipy.stats import expon

lambda_val = 2  # Rate parameter ()
size = 1000  # Number of samples

# Generate random samples
samples = np.random.exponential(1/lambda_val, size)

# Plot histogram of generated data
plt.hist(samples, bins=50, density=True, alpha=0.6, color='b', edgecolor='black')

# Plot the theoretical PDF
x = np.linspace(0, max(samples), 100)
pdf = lambda_val * np.exp(-lambda_val * x)
plt.plot(x, pdf, 'r', lw=2, label=f'={lambda_val}')

plt.xlabel("Time Between Events")
plt.ylabel("Probability Density")
plt.title("Exponential Distribution")
plt.legend()
plt.show()
```

```python
data = np.array([1, 2, 3, 4, 5])
mean = np.mean(data)
variance = np.var(data)
cov_matrix = np.cov(data, data)

print("Mean:", mean)
print("Variance:", variance)
print("Covariance Matrix:\n", cov_matrix)
```

```python
import sympy as sp

x = sp.Symbol('x')
f = x**2
derivative = sp.diff(f, x)

print("Derivative of x^2:", derivative)
```

```python
integral = sp.integrate(x**2, x)

print("Integral of x^2:", integral)
```

```python
def gradient_descent(learning_rate=0.1, epochs=100):
    x = 10  # Initial value
    for _ in range(epochs):
        grad = 2 * x  # Derivative of x^2
        x = x - learning_rate * grad
    return x

optimized_x = gradient_descent()
print("Optimized x:", optimized_x)
```
